# N-Queens Performance Analysis

This notebook reads the `results.csv` file generated by `benchmark_all.py` and visualizes the performance differences between the Python, C++, and Java implementations.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
sns.set_theme(style="whitegrid")
```

```python
# Load the dataset
try:
    df = pd.read_csv('results.csv')
except FileNotFoundError:
    print("Error: results.csv not found. Please run benchmark_all.py first.")

df['Time (s)'] = pd.to_numeric(df['Time (s)'], errors='coerce')
df = df.dropna()

color_map = {
    'Python': 'background-color: #a8d5e2; color: black;',   
    'C++': 'background-color: #f2b6b6; color: black;',      
    'Java': 'background-color: #f6d7a7; color: black;'
}

def highlight_rows_by_language(row):
    style = color_map.get(row['Language'], '')
    return [style] * len(row)

df.style.apply(highlight_rows_by_language, axis=1)
```

## Execution Time vs. Board Size (Linear Scale)

This first chart shows the raw execution time. As you can see, the exponential nature of the backtracking algorithm makes the non-bitboard solutions quickly dominate the chart, hiding the performance of the faster implementations.

```python
plt.figure(figsize=(12, 8))

sns.lineplot(
    data=df,
    x='N',
    y='Time (s)',
    hue='Runner',
    style='Runner',
    markers=True,
    dashes=False,
    lw=2
)

plt.title('Execution Time vs. Board Size (N)', fontsize=16)
plt.xlabel('Board Size (N)', fontsize=12)
plt.ylabel('Time (s)', fontsize=12)
plt.legend(title='Implementation')
plt.xticks(df['N'].unique())
plt.show()
```

## Execution Time vs. Board Size (Logarithmic Scale)

A logarithmic scale on the y-axis is essential for comparing algorithms with exponential complexity. This chart clearly shows the massive performance gap between the C++ (Bitboard) version and all other implementations. It also makes the differences between Python, Java, and standard C++ visible.

```python
plt.figure(figsize=(12, 8))

sns.lineplot(
    data=df,
    x='N',
    y='Time (s)',
    hue='Runner',
    style='Runner',
    markers=True,
    dashes=False,
    lw=2
)

plt.yscale('log') # Set y-axis to log scale

plt.title('Execution Time vs. Board Size (N) - Log Scale', fontsize=16)
plt.xlabel('Board Size (N)', fontsize=12)
plt.ylabel('Time (s) (Log Scale)', fontsize=12)
plt.legend(title='Implementation')
plt.xticks(df['N'].unique())
plt.grid(True, which="both", ls="--")
plt.show()
```

## Analysis & Takeaways

Based on the log-scale chart, we can draw clear conclusions:

1.  **Algorithm > Language:** The C++ (Bitboard) solution is orders of magnitude faster than any other implementation. This demonstrates that the choice of algorithm and data structure (using bit-manipulation for $O(1)$ set-wise operations) has a far greater impact on performance than the choice of language.

2.  **Compiled vs. Interpreted:** As expected, the compiled C++ (Standard) implementation is significantly faster than the Python (Standard) implementation, even though they use the exact same backtracking logic (helper arrays).

3.  **Java's Position:** The Java implementation places itself between C++ and Python. Its JIT (Just-In-Time) compiler makes it much faster than CPython, but it still has overhead (like JVM startup, bounds checking, and object management) that makes it slower than the ahead-of-time compiled C++.

4.  **Expressiveness:**
    * **Python** is arguably the most expressive and readable, with the least amount of boilerplate code.
    * **Java** is more verbose due to its class-based, object-oriented structure, but is very clear and maintainable.
    * **C++ (Bitboard)** is the least expressive. The bit-shifting logic (`(ld | bit) << 1`) is clever and highly optimized but is completely opaque to anyone not familiar with the bitboard technique.
